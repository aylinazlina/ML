# Separate X,Y

```python
x = df.drop('diagnosis',axis=1)
x.head()
```

```python
y = df[['diagnosis']]
y.head()
```

# Split Train & Test

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest = tts(x,y,test_size=0.3,random_state=2)
```

```python
xtrain.shape
```

```python
xtest
```

```python
ytest
```

```python
xtrain
```

```python
ytrain
```

# Applying KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
ne = KNeighborsClassifier()
```

```python
ne.fit(xtrain,ytrain)
```

```python
 ne.predict([[18.94, 21.31, 123.6, 1130, 0.09009, 0.1029, 0.108, 0.07951, 0.1582, 0.05461, 0.7888, 0.7975, 5.486, 96.05, 0.004444, 0.01652, 0.02269, 0.0137, 0.01386, 0.001698, 24.86, 26.58, 165.9, 1866, 0.1193, 0.2336, 0.2687, 0.1789, 0.2551, 0.06589]])

```

```python
pred = ne.predict(xtest)
```

```python
pred
```

# Evaluate Model

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, RocCurveDisplay
```

```python
print(classification_report(ytest,pred))
```

```python
con_matrix = confusion_matrix(ytest,pred)
```

```python
con_matrix
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(con_matrix,annot=True)
plt.xlabel('Predicted Result')
plt.ylabel('Actual value')
```

```python
accuracy_score(ytest,pred)
```

```python
RocCurveDisplay.from_estimator(ne, xtest, ytest)
```
